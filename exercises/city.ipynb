```python
df = df[df.Temp > -60]
```

```python
df = df[df.Temp > -60]
df["Dayofyear"] = df["Date"].astype("datetime64").dt.dayofyear
```

```python
df_israel = df[df["Country"] == "Israel"]
```

```python
df_israel.plot(kind='scatter', x='Dayofyear', y='Temp', c="Year", colormap='Paired')    
```
